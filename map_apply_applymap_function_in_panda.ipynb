```python
import pandas as pd
```

```python
data={
    'math':[40,45,68,90,44],
    'english':[90,55,38,30,80],
    'hindi':[77,80,66,20,90],
    'science':[30,45,98,50,35]}

df=pd.DataFrame(data)
df
```

```python
#map function
def avg(row):
    return (row['math'] + row['english'] + row['hindi'] + row['science']) / 4
df['avg']=df.apply(avg,axis=1)
df


```

```python
df_change={'math':'Physices','english':'Biology','hindi':'chemistry','science':'Gernal_Knowledge'}
df.rename(columns=df_change,inplace=True) # passing the column is mendatry
df
```

```python
import pandas as pd

# Example DataFrame
data = {'Physics': [40, 45, 68, 90, 44]}
df3 = pd.DataFrame(data)

# Mapping dictionary
add_5_to_physics = {40: 45, 45: 50, 68: 74, 90: 95, 44: 49}

# Adding a new column based on mapping
df3['new_marks'] = df3['Physics'].map(add_5_to_physics)

print(df3)
```

```python
df2=pd.DataFrame({
    'month_name':['jan','feb','mar','jun']
})
cap={'jan':'JAN','feb':'FEB','mar':'MAR','jun':'JUN'}

df2['capital']=df2['month_name'].map(cap)
df2
```
