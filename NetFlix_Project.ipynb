```python
import numpy as np
import pandas as pd
import plotly.express as px
from textblob import   TextBlob
```

```python
df=pd.read_csv("/content/netflix_titles.csv")
df.head()
```

```python
df.columns
```

```python
x=df.groupby(["rating"]).size().reset_index(name="count")
x
```

```python
pichart=px.pie(x,values='count',names="rating",title="Distribution Of Rating Count content in Netflix")
pichart.show()
```

```python
df["director"]=df["director"].fillna('Director Not Specified')
df
```

```python
Director_list=pd.DataFrame()
print(Director_list)
```

```python
Director_list=df["director"].str.split(",",expand=True).stack()
print(Director_list)
```

```python
Director_list = df['director'].str.split(',', expand=True).stack().reset_index(drop=True).to_frame(name="Director")
print(Director_list)
```

```python
Director_list.columns
```

```python
directors=Director_list.groupby(['Director']).size().reset_index(name='Total_count')
print(directors)
```

```python
directors = directors[directors['Director'] != "Director Not Specified"]
print(directors)
```

```python
directors = directors.sort_values(by=['Total_count'],ascending=False)
print(directors)
```

```python
Top5Directors=directors.head()
Top5Directors
```

```python
data=Top5Directors.sort_values(by=['Total_count'],ascending=True)
barchart=px.bar(data,x='Total_count',y='Director',title='Top 5 Director')
barchart.show()
```

```python
df['cast']=df['cast'].fillna("Not Specified")
```

```python
cast_df=pd.DataFrame()
print(cast_df)
```

```python
cast_df = df['cast'].str.split(',', expand=True).stack().reset_index(drop=True).to_frame(name="Actor")
cast_df
```

```python
actors=cast_df.groupby(['Actor']).size().reset_index(name='Total Count')
actors
```

```python
actors = actors[actors['Actor'] != "Not Specified"]
actors
```

```python
actors=actors.sort_values(by=['Total Count'],ascending=False)
Top5Actor=actors.head()
data=Top5Actor.sort_values(by=['Total Count'],ascending=True)
actor_data=px.bar(data,x='Total Count',y='Actor',title="Top 5 Actors")
actor_data.show()
```
